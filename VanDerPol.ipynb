```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import time
```

```python
# Define the Van Der Pol oscillator equations
def van_der_pol(t, y, mu):
    y1, y2 = y
    dy1 = y2
    dy2 = mu * ((1 - y1**2) * y2) - y1
    return [dy1, dy2]
```

```python
# Define the initial conditions and parameters
y0 = [2.0, 0.0]
tmax = 1000.0
mus = [10, 100, 1000]
atols = [1e-6, 1e-9, 1e-12]
```

## RK45 method
The following three cells of code corresponds to mu=10, 100, 100 respectively.

```python
# RK45 with mu = 10
mu = mus[0]

for atol in atols:
    # Solve the ODE using RK45 solver
    start_time = time.time()
    solution_rk45 = solve_ivp(van_der_pol, [0, tmax], y0, args=(mu,), method='RK45', atol=atol, rtol=atol)
    rk45_time = time.time() - start_time

    # Plot the solution on the phase plane
    plt.subplot(2, 1, 1)
    plt.plot(solution_rk45.y[0], solution_rk45.y[1], label=f'Atol={atol}')
    plt.title(f'RK45 Solver (Mu={mu})')
    plt.xlabel('y1')
    plt.ylabel('y2')
    plt.legend()

    # Plot log(error) versus log(CPU time)
    plt.subplot(2, 1, 2)
    plt.loglog(atol, rk45_time, 'o', label=f'RK45 Atol={atol}')
    plt.title('Log(Error) versus Log(CPU Time)')
    plt.xlabel('Log(Error)')
    plt.ylabel('Log(CPU Time)')
    plt.legend()

plt.tight_layout()
```
